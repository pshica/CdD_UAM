# Cálculo de Información Mutua entre dos series temporales

**Introducción**

El cálculo de Información Mutua (IM) es una herramienta estadística fundamental para medir la dependencia entre dos series temporales y cuantificar la cantidad de información compartida entre ellas. En este trabajo, aplicaremos este método para analizar la relación entre las series temporales de actividad de dos neuronas, VD y LP, bajo diferentes condiciones experimentales. 

La IM se utilizará para evaluar el flujo de información entre las neuronas en distintos estados del sistema (control, inyección de GABA y recuperación). Dado que la actividad neuronal se expresa en eventos de *spikes*, estos se considerarán como los principales portadores de información, permitiendo calcular la IM a partir de la probabilidad de ocurrencia de dichos eventos en diversas configuraciones.

Los resultados obtenidos serán representados en función de la ventana de resolución y permitirán observar cómo varía la IM en función de los diferentes estados y configuraciones experimentales. La interpretación de estos resultados proporcionará una visión profunda sobre la conectividad funcional entre las neuronas estudiadas y cómo la inyección de GABA o la recuperación del sistema afectan la transferencia de información en el sistema neuronal.

# PROCEDIMIENTOS PARA TRATAR LA INFORMACION
**Cálculo de la Información Mutua (IM) para Palabras de 1 a 8 Bits**

Para calcular la Información Mutua (IM) entre dos series temporales $ X = \{x_i\} $ y $ Y = \{y_i\} $, definimos la IM entre dos eventos $ x_i $ y $ y_i $ como:

$$
I(x_i; y_i) = \log_2 \left( \frac{p(x_i | y_i)}{p(x_i)} \right)
$$

Donde:
- $ p(x_i | y_i) $ es la probabilidad condicional de $ x_i $ dado $ y_i $.
- $ p(x_i) $ es la probabilidad marginal de $ x_i $.

La Información Mutua promedio o media, que se denota como $ MI(x, y) $, se calcula como la esperanza matemática de la IM para todos los posibles pares de eventos en $ X $ y $ Y $:

$$
MI(x, y) = \sum_{x, y} p(x, y) \, I(x; y)
$$

Donde:
- $ p(x, y) $ es la probabilidad conjunta de los eventos $ x $ y $ y $.
- $ I(x; y) $ representa la IM entre los eventos $ x $ y $ y $, como se definió anteriormente.

Esta IM promedio representa la cantidad de información que contiene la variable $ x $ acerca de la variable $ y $ o viceversa. En otras palabras, es la intersección de la entropía de $ x $ con to Temporal de la Señal

Para el tratamiento temporal de la señal, consideramos una secuencia discreta de eventos temporales $ \{e_t, t=1, \ldots, N\} $, donde cada evento $ e_t $ puede tomar el valor de 0 o 1 (representando la ocurrencia o ausencia de un *spike*).

Definimos una palabra $ W_t^L $ de tamaño $ L $ en el tiempo $ t $ como la secuencia de símbolos:

$$
W_t^L = \{e_t, e_{t+1}, \ldots, e_{t+L-1}\}
$$

De esta manera, cada palabra $ W_t^L $ representa una ventana de $ L $ instantes consecutivos en la serie temporal, donde $ L $ varía de 1 a 8 bits en nuestro caso de estudio.

Por lo tanto, en una serie temporal de longitud $ N $, tendremos $ N - L + 1 $ palabras de longitud $ L ua para Palabras de Longitud Variable

Para calcular la Información Mutua (IM) utilizando palabras de longitud de 1 a 8 bits:

1. **Construcción de Palabras**: Extraemos palabras de longitud $ L $ bits (donde $ L = 1, 2, \ldots, 8 $) para ambas series temporales $ X $ y $ Y $.

2. **Cálculo de Probabilidades**:
   - Calculamos la probabilidad conjunta $ p(W_x^L, W_y^L) $ de cada par de palabras $ W_x^L $ y $ W_y^L $.
   - Calculamos las probabilidades marginales $ p(W_x^L) $ y $ p(W_y^L) $ para cada serie.

3. **Cálculo de la Información Mutua (IM)**:
   - La IM entre cada par de palabras $ W_x^L $ y $ W_y^L $ se calcula como:

   $$
   I(W_x^L; W_y^L) = \log_2 \left( \frac{p(W_x^L | W_y^L)}{p(W_x^L)} \right)
   $$

4. **IM Promedio**: Finalmente, la IM promedio entre $ X $ y $ Y $ se obtiene sumando sobre todas las palabras posibles:

   $$
   MI(X, Y) = \sum_{W_x^L, W_y^L} p(W_x^L, W_y^L) \, I(W_x^L; W_y^L)
   $$

Este cálculo se repite para cada longitud de palabra desde 1 hasta 8 bits para obtener la IM en función del tamaño de las palabras.

**El cálculo de la transferencia de entropía normalizada de la neurona VD a LP y viceversa.**


Para medir la transferencia de información entre las neuronas VD (como neurona de entrada) y LP (como neurona de salida, o respuesta), utilizamos el concepto de Información Mutua (IM) y su versión normalizada, llamada entropía normalizada.

1. **Definición de la Transferencia de Información:**

   Para calcular la transferencia de información, consideramos la probabilidad conjunta $P_{RS}(r_i, s_i)$, donde:
   - $r_i$ representa las respuestas de la neurona de salida (LP).
   - $s_i$ representa los estímulos de la neurona de entrada (VD).

   Esta probabilidad nos permite estimar el promedio de la información transmitida a través de una secuencia de palabras formadas por la respuesta a un estímulo dado. Este promedio se calcula mediante datos experimentales o simulaciones disponibles.

2. **Entropía Normalizada:**

   La medida de información se puede normalizar utilizando la información mutua normalizada o entropía normalizada $E_{RS}$, que se define como:

   $$
   E_{RS} = \frac{I_{RS}}{H(S)}
   $$

   Donde:
   - $I_{RS}$ es la información mutua entre el estímulo $S$ y la respuesta $R$.
   - $H(S)$ es la entropía del estímulo, que representa la máxima cantidad de información que el estímulo puede transmitir a la respuesta.

   La entropía normalizada $E_{RS}$ toma valores en el rango $0 \leq E_{RS} \leq 1$:
   - Si $E_{RS} = 0$, significa que toda la información es perdida; es decir, el estímulo y la respuesta son completamente independientes.
   - Si $E_{RS} = 1$, indica que hay una sincronización completa entre el estímulo y la respuesta de la neurona, reflejando un seguimiento perfecto del estímulo por parte de la neurona respuesta.

3. **Cálculo para Diferentes Longitudes de Palabras:**

   Para calcular la transferencia de entropía usando palabras de diferentes longitudes (de 1 a 8 bits), seguimos estos pasos:

   - Construimos palabras de longitud $L$ bits de las series temporales de ambas neuronas, donde $L = 1, 2, \ldots, 8$.
   - Calculamos la probabilidad conjunta $P_{RS}(r_i, s_i)$ para cada par de palabras de entrada y respuesta.
   - Usamos la fórmula de entropía normalizada para cada longitud de palabra, obteniendo $E_{RS}$ para cada valor de $L$.

Esta metodología permite evaluar cómo cambia la transferencia de información entre VD y LP en función de la longitud de las palabras.


**Declaración de librerías y algunas funciones principales**

```python
import pandas as pd
import numpy as np
import math
from os import listdir
import matplotlib.pyplot as plt
from scipy.signal import find_peaks
from collections import Counter
from itertools import permutations

###########Función optimizada para detectar spikes sin usar for########
def spikes(signal, threshold):
    # Crear una serie binaria donde 1 indica que el valor supera el umbral
    above_threshold = signal >= threshold

    # Encontrar los cambios donde pasa de estar por debajo a por encima del umbral y viceversa
    transitions = np.diff(above_threshold.astype(int))

    # Índices donde hay cruces hacia arriba (de 0 a 1) y hacia abajo (de 1 a 0)
    spike_start = np.where(transitions == 1)[0] + 1  # Cruce hacia arriba
    spike_end = np.where(transitions == -1)[0] + 1   # Cruce hacia abajo
    
    # Asegurarse de que hay un par de inicio-fin de spike válido
    if len(spike_start) > len(spike_end):
        spike_start = spike_start[:len(spike_end)]  # Si hay más inicios que finales, recortamos el exceso

    # Calcular el punto medio entre cada cruce hacia arriba y su correspondiente cruce hacia abajo
    spike_midpoints = (spike_start + spike_end) // 2

    # Crear una serie binaria donde 1 indica la posición de los spikes
    binary_spikes = np.zeros(len(signal))
    binary_spikes[spike_midpoints] = 1  # Colocar 1 en el punto medio de cada spike
    return binary_spikes
    
def crear_palabras_de_bits(binary_signal, n_bits):
    palabras = []
    for i in range(len(binary_signal) - n_bits + 1):
        palabra = ''.join(map(str, binary_signal[i:i + n_bits]))
        palabras.append(palabra)
    return palabras

# Función para binarizar la señal en función de la ventana y el umbral
def binarizar_signal(signal, window_size, threshold=0.4):
    return np.array([1 if np.max(signal[i:i + window_size]) > threshold else 0 
                     for i in range(0, len(signal), window_size)])
    
###########Cálculo de la información mutua#############
def calcular_informacion_mutua(palabras_lp, palabras_vd):
    # Calcular probabilidades conjuntas
    conjunto_palabras = list(zip(palabras_lp, palabras_vd))
    prob_conjunta = Counter(conjunto_palabras)
    total_palabras = len(conjunto_palabras)
    for k in prob_conjunta:
        prob_conjunta[k] /= total_palabras  # Dividir cada frecuencia entre el total para obtener la probabilidad conjunta

    # Calcular probabilidades marginales para LP y VD
    prob_lp = Counter(palabras_lp)
    prob_vd = Counter(palabras_vd)
    for k in prob_lp:
        prob_lp[k] /= total_palabras  # Probabilidad marginal de cada palabra en LP
    for k in prob_vd:
        prob_vd[k] /= total_palabras  # Probabilidad marginal de cada palabra en VD

    # Calcular la información mutua
    IM = 0
    for (x, y), p_xy in prob_conjunta.items():
        p_x = prob_lp[x]
        p_y = prob_vd[y]
        if p_xy > 0:  # Solo considerar valores de probabilidad conjunta mayores a 0
            IM += p_xy * np.log2(p_xy / (p_x * p_y))
    
    return IM

############ Función para calcular la entropía de una serie de palabras##########
def calcular_entropia(palabras):
    prob_palabras = Counter(palabras)
    total_palabras = len(palabras)
    entropia = 0
    for count in prob_palabras.values():
        p = count / total_palabras
        entropia -= p * np.log2(p)
    return entropia
    
############# Graficar la serie temporal ###################
def plot_signals(data, interval_range):
    """
    Función para graficar las señales LP y VD en un intervalo específico.
    
    Parámetros:
    - data: DataFrame con las señales, debe contener las columnas 'LP' y 'VD'.
    - interval_range: Lista o tupla con los límites [inicio, fin] del intervalo a graficar (en milisegundos).
    """
    # Definir colores para LP y VD
    colors = plt.cm.viridis(np.linspace(0, 1, 2))  # Solo necesitamos dos colores

    # Crear la figura y los ejes
    fig, axs = plt.subplots(2, figsize=(15, 8), sharex=True)
    fig.suptitle(f'Data from {interval_range[0]} to {interval_range[1]} milliseconds', fontsize=16)

    # Graficar la serie temporal de LP
    axs[0].plot(np.arange(interval_range[0], interval_range[1]), 
                data.iloc[interval_range[0]:interval_range[1]]["LP"], 
                color=colors[0], linewidth=1, label="LP")
    axs[0].set_ylabel("LP", fontsize=12)
    axs[0].legend(loc="upper right")
    axs[0].grid(True)

    # Graficar la serie temporal de VD
    axs[1].plot(np.arange(interval_range[0], interval_range[1]), 
                data.iloc[interval_range[0]:interval_range[1]]["VD"], 
                color=colors[1], linewidth=1, label="VD")
    axs[1].set_ylabel("VD", fontsize=12)
    axs[1].legend(loc="upper right")
    axs[1].grid(True)

    # Configurar el eje x
    axs[1].set_xlabel('Time (milliseconds)', fontsize=12)
    plt.tight_layout(rect=[0, 0, 1, 0.95])  # Ajustar layout para dar espacio al título
    plt.show()
    
################  CALCULA PATRONES ORDINALES ###################
################################################################

def calcular_patrones_ordinales(trozo, columna, orden=4, tau=2):
    """
    Calcula los patrones ordinales para una columna específica de un DataFrame dado (trozo).
    
    Parámetros:
    - trozo: DataFrame que contiene la señal de interés.
    - columna: Nombre de la columna de la señal (como 'LP' o 'VD').
    - orden: Longitud del patrón ordinal (d).
    - tau: Paso entre puntos sucesivos en el patrón ordinal (τ).
    
    Retorna:
    - Lista de patrones ordinales obtenidos para la señal especificada.
    """
    # Seleccionar la señal
    signal = trozo[columna]

    # Generar los patrones ordinales
    patrones_ordinales = []
    for i in range(len(signal) - (orden - 1) * tau):
        # Extraer el segmento con el espaciado de tau
        segmento = signal[i:i + orden * tau:tau]
        orden_indices = np.argsort(segmento)
        patron = tuple(orden_indices)
        patrones_ordinales.append(patron)
    
    return patrones_ordinales

# Función para calcular probabilidades de patrones ordinales
def calcular_probabilidades_patrones_ordinales(patrones_ordinales, orden):
    """
    Calcula las probabilidades de los patrones ordinales.

    Parámetros:
    - patrones_ordinales: Lista de patrones ordinales.
    - orden: Longitud del patrón ordinal (d).
    
    Retorna:
    - Un diccionario con las probabilidades de cada patrón ordinal.
    """
    # Generar todas las permutaciones posibles para los patrones ordinales
    patrones = list(permutations(range(orden)))

    # Contar la frecuencia de cada patrón
    frecuencias = {patron: patrones_ordinales.count(patron) for patron in patrones}

    # Calcular la probabilidad de cada patrón
    total_patrones = len(patrones_ordinales)
    probabilidades = {patron: frecuencia / total_patrones for patron, frecuencia in frecuencias.items()}

    return probabilidades

# Función para calcular la probabilidad conjunta de patrones ordinales entre dos señales
def calcular_probabilidad_conjunta(patrones_lp, patrones_vd):
    """
    Calcula la probabilidad conjunta de los patrones ordinales entre dos señales.

    Parámetros:
    - patrones_lp: Lista de patrones ordinales para la primera señal (ej., "LP").
    - patrones_vd: Lista de patrones ordinales para la segunda señal (ej., "VD").

    Retorna:
    - Un diccionario con las probabilidades conjuntas de cada par de patrones.
    """
    patrones_conjuntos = list(zip(patrones_lp, patrones_vd))
    frecuencias_conjuntas = {pair: patrones_conjuntos.count(pair) for pair in set(patrones_conjuntos)}
    total_conjuntos = len(patrones_conjuntos)
    probabilidades_conjuntas = {pair: freq / total_conjuntos for pair, freq in frecuencias_conjuntas.items()}
    return probabilidades_conjuntas

# Función para calcular la información mutua ordinal entre dos señales
def calcular_informacion_mutua_ordinal(probabilidades_lp, probabilidades_vd, probabilidades_conjuntas):
    """
    Calcula la información mutua ordinal entre dos señales.

    Parámetros:
    - probabilidades_lp: Diccionario con las probabilidades de los patrones ordinales de la primera señal.
    - probabilidades_vd: Diccionario con las probabilidades de los patrones ordinales de la segunda señal.
    - probabilidades_conjuntas: Diccionario con las probabilidades conjuntas de los pares de patrones.

    Retorna:
    - El valor de la información mutua ordinal entre las dos señales.
    """
    informacion_mutua = 0.0
    for (patron_lp, patron_vd), prob_conjunta in probabilidades_conjuntas.items():
        prob_lp = probabilidades_lp.get(patron_lp, 0)
        prob_vd = probabilidades_vd.get(patron_vd, 0)
        if prob_lp > 0 and prob_vd > 0 and prob_conjunta > 0:
            informacion_mutua += prob_conjunta * np.log(prob_conjunta / (prob_lp * prob_vd))
    return informacion_mutua

# Función principal para calcular la información mutua entre dos señales de un trozo
def calcular_informacion_mutua_trozos(trozo, orden=4, tau=2):
    """
    Calcula la información mutua ordinal entre dos señales de un trozo específico.

    Parámetros:
    - trozo: DataFrame que contiene las señales.
    - orden: Longitud del patrón ordinal (d).
    - tau: Paso entre puntos sucesivos en el patrón ordinal (τ).

    Retorna:
    - La información mutua ordinal entre las señales "LP" y "VD" del trozo.
    """
    # Calcular los patrones ordinales
    patrones_ordinales_lp = calcular_patrones_ordinales(trozo, "LP", orden, tau)
    patrones_ordinales_vd = calcular_patrones_ordinales(trozo, "VD", orden, tau)

    # Calcular las probabilidades de los patrones ordinales
    probabilidades_lp = calcular_probabilidades_patrones_ordinales(patrones_ordinales_lp, orden)
    probabilidades_vd = calcular_probabilidades_patrones_ordinales(patrones_ordinales_vd, orden)

    # Calcular la probabilidad conjunta
    probabilidades_conjuntas = calcular_probabilidad_conjunta(patrones_ordinales_lp, patrones_ordinales_vd)

    # Calcular la información mutua ordinal
    informacion_mutua = calcular_informacion_mutua_ordinal(probabilidades_lp, probabilidades_vd, probabilidades_conjuntas)
    
    return informacion_mutua

############Grafica los patrones ordinales para las señales especificadas de un trozo dado.###########
def graficar_patrones_ordinales(trozo, señales, orden=4, tau=2, num_muestras=50):
    """
    Grafica los patrones ordinales para las señales especificadas de un trozo dado.
    
    Parámetros:
    - trozo: DataFrame que contiene las señales de interés.
    - señales: Lista de nombres de columnas de las señales a graficar (por ejemplo, ["LP", "VD"]).
    - orden: Longitud del patrón ordinal (d).
    - tau: Paso entre puntos sucesivos en el patrón ordinal (τ).
    - num_muestras: Número de muestras iniciales de la señal para graficar.
    """
    # Función para obtener el patrón ordinal de un segmento
    def obtener_patron_ordinal(segmento):
        orden_indices = np.argsort(segmento)
        return tuple(orden_indices)

    # Iterar sobre cada señal en la lista
    for señal in señales:
        signal = trozo[señal]
        patrones_ordinales = []

        # Calcular los patrones ordinales
        for i in range(len(signal) - (orden - 1) * tau):
            segmento = signal[i:i + orden * tau:tau]
            patron = obtener_patron_ordinal(segmento)
            patrones_ordinales.append(patron)

        # Graficar la señal y sus patrones ordinales
        fig, ax = plt.subplots(figsize=(10, 5))
        ax.plot(signal[:num_muestras], color="grey", alpha=0.5, label=f"Serie Temporal {señal}")
        for i in range(0, num_muestras - (orden - 1) * tau, tau):
            segmento = signal[i:i + orden * tau:tau]
            patron = obtener_patron_ordinal(segmento)
            ax.plot(range(i, i + orden * tau, tau), segmento, marker="o", linestyle="-", linewidth=2, markersize=8,
                    label=f"Patrón {patron}" if i == 0 else "")
        
        # Configuración del gráfico
        ax.set_title(f"TROZO - Patrones Ordinales de {señal}")
        ax.set_xlabel("Tiempo")
        ax.set_ylabel(f"Valor de {señal}")
        ax.legend()
        plt.show()
```

**Calcula la IM 1 bits hasta 8 bits**

```python
# Rango de tamaños de ventana en segundos (1 a 150 segundos)#######################
window_sizes_seconds = range(1, 150)
sampling_interval = 0.1  # Intervalo de muestreo en segundos

# Diccionario para almacenar los resultados de IM para cada longitud de palabra
im_results = {f"{n}_bits": [] for n in range(1, 9)}

# Calcular la IM para cada tamaño de ventana y cada longitud de palabra
for window_seconds in window_sizes_seconds:
    window_size_samples = int(window_seconds / sampling_interval)  # Convertir a número de muestras

    # Binarizar las señales LP y VD para el tamaño de ventana actual
    binary_signal_lp = binarizar_signal(signal_lp, window_size_samples)
    binary_signal_vd = binarizar_signal(signal_vd, window_size_samples)

    # Calcular la IM para cada longitud de palabra de 1 a 8 bits
    for n in range(1, 9):
        palabras_lp = crear_palabras_de_bits(binary_signal_lp, n)
        palabras_vd = crear_palabras_de_bits(binary_signal_vd, n)
        
        # Calcular la IM para la longitud de palabra actual
        IM_n_bits = calcular_informacion_mutua(palabras_lp, palabras_vd)
        im_results[f"{n}_bits"].append(IM_n_bits)
```

**Entropia normalizada de VD a LP y de LP a VP**

```python
# Diccionarios para almacenar los resultados de Entropía Normalizada para cada longitud de palabra en ambas direcciones
entropia_normalizada_vd_lp = {f"{n}_bits": [] for n in range(1, 9)}
entropia_normalizada_lp_vd = {f"{n}_bits": [] for n in range(1, 9)}

# Calcular la Entropía Normalizada para cada tamaño de ventana y cada longitud de palabra en ambas direcciones
for window_seconds in window_sizes_seconds:
    window_size_samples = int(window_seconds / sampling_interval)  # Convertir a número de muestras

    # Binarizar ambas señales para el tamaño de ventana actual
    binary_signal_vd = binarizar_signal(signal_vd, window_size_samples)
    binary_signal_lp = binarizar_signal(signal_lp, window_size_samples)

    for n in range(1, 9):
        # Crear palabras de n bits para ambas señales
        palabras_vd = crear_palabras_de_bits(binary_signal_vd, n)
        palabras_lp = crear_palabras_de_bits(binary_signal_lp, n)

        # Entropía Normalizada de VD a LP
        IM_vd_lp = calcular_informacion_mutua(palabras_vd, palabras_lp)
        H_vd = calcular_entropia(palabras_vd)
        E_vd_lp = IM_vd_lp / H_vd if H_vd > 0 else 0
        entropia_normalizada_vd_lp[f"{n}_bits"].append(E_vd_lp)

        # Entropía Normalizada de LP a VD
        IM_lp_vd = calcular_informacion_mutua(palabras_lp, palabras_vd)
        H_lp = calcular_entropia(palabras_lp)
        E_lp_vd = IM_lp_vd / H_lp if H_lp > 0 else 0
        entropia_normalizada_lp_vd[f"{n}_bits"].append(E_lp_vd)
```

```python
# Crear gráficos en bloques de 1 fila por 2 columnas para mostrar la Entropía Normalizada en ambas direcciones
fig, axs = plt.subplots(1, 2, figsize=(16, 4))  # Crear una cuadrícula de 1 fila por 2 columnas

# Gráfico 1: Entropía Normalizada de VD a LP
for n in range(1, 9):
    axs[0].plot(window_sizes_seconds, entropia_normalizada_vd_lp[f"{n}_bits"], label=f"{n} bits")
axs[0].set_xlabel("Tamaño de ventana (segundos)", fontsize=8)
axs[0].set_ylabel("Entropía Normalizada (E)", fontsize=8)
axs[0].set_title("Entropía Normalizada de VD a LP", fontsize=10)
axs[0].legend(title="Longitud de palabra", fontsize=6)
axs[0].grid()

# Gráfico 2: Entropía Normalizada de LP a VD
for n in range(1, 9):
    axs[1].plot(window_sizes_seconds, entropia_normalizada_lp_vd[f"{n}_bits"], label=f"{n} bits")
axs[1].set_xlabel("Tamaño de ventana (segundos)", fontsize=8)
axs[1].set_ylabel("Entropía Normalizada (E)", fontsize=8)
axs[1].set_title("Entropía Normalizada de LP a VD", fontsize=10)
axs[1].legend(title="Longitud de palabra", fontsize=6)
axs[1].grid()

plt.tight_layout()  # Ajustar espaciado para que los subgráficos no se sobrepongan
plt.show()
```

**Calcula la IM 1 bits hasta 8 bits**

```python
# Rango de tamaños de ventana en segundos (1 a 120 segundos)#######################
window_sizes_seconds = range(1, 120)
sampling_interval = 0.1  # Intervalo de muestreo en segundos

# Diccionario para almacenar los resultados de IM para cada longitud de palabra
im_results = {f"{n}_bits": [] for n in range(1, 9)}

# Calcular la IM para cada tamaño de ventana y cada longitud de palabra
for window_seconds in window_sizes_seconds:
    window_size_samples = int(window_seconds / sampling_interval)  # Convertir a número de muestras

    # Binarizar las señales LP y VD para el tamaño de ventana actual
    binary_signal_lp = binarizar_signal(signal_lp, window_size_samples)
    binary_signal_vd = binarizar_signal(signal_vd, window_size_samples)

    # Calcular la IM para cada longitud de palabra de 1 a 8 bits
    for n in range(1, 9):
        palabras_lp = crear_palabras_de_bits(binary_signal_lp, n)
        palabras_vd = crear_palabras_de_bits(binary_signal_vd, n)
        
        # Calcular la IM para la longitud de palabra actual
        IM_n_bits = calcular_informacion_mutua(palabras_lp, palabras_vd)
        im_results[f"{n}_bits"].append(IM_n_bits)
```

**Entropia normalizada de VD a LP y de LP a VP**

```python
# Diccionarios para almacenar los resultados de Entropía Normalizada para cada longitud de palabra en ambas direcciones
entropia_normalizada_vd_lp = {f"{n}_bits": [] for n in range(1, 9)}
entropia_normalizada_lp_vd = {f"{n}_bits": [] for n in range(1, 9)}

# Calcular la Entropía Normalizada para cada tamaño de ventana y cada longitud de palabra en ambas direcciones
for window_seconds in window_sizes_seconds:
    window_size_samples = int(window_seconds / sampling_interval)  # Convertir a número de muestras

    # Binarizar ambas señales para el tamaño de ventana actual
    binary_signal_vd = binarizar_signal(signal_vd, window_size_samples)
    binary_signal_lp = binarizar_signal(signal_lp, window_size_samples)

    for n in range(1, 9):
        # Crear palabras de n bits para ambas señales
        palabras_vd = crear_palabras_de_bits(binary_signal_vd, n)
        palabras_lp = crear_palabras_de_bits(binary_signal_lp, n)

        # Entropía Normalizada de VD a LP
        IM_vd_lp = calcular_informacion_mutua(palabras_vd, palabras_lp)
        H_vd = calcular_entropia(palabras_vd)
        E_vd_lp = IM_vd_lp / H_vd if H_vd > 0 else 0
        entropia_normalizada_vd_lp[f"{n}_bits"].append(E_vd_lp)

        # Entropía Normalizada de LP a VD
        IM_lp_vd = calcular_informacion_mutua(palabras_lp, palabras_vd)
        H_lp = calcular_entropia(palabras_lp)
        E_lp_vd = IM_lp_vd / H_lp if H_lp > 0 else 0
        entropia_normalizada_lp_vd[f"{n}_bits"].append(E_lp_vd)
```

```python
# Crear gráficos en bloques de 1 fila por 2 columnas para mostrar la Entropía Normalizada en ambas direcciones
fig, axs = plt.subplots(1, 2, figsize=(16, 4))  # Crear una cuadrícula de 1 fila por 2 columnas

# Gráfico 1: Entropía Normalizada de VD a LP
for n in range(1, 9):
    axs[0].plot(window_sizes_seconds, entropia_normalizada_vd_lp[f"{n}_bits"], label=f"{n} bits")
axs[0].set_xlabel("Tamaño de ventana (segundos)", fontsize=8)
axs[0].set_ylabel("Entropía Normalizada (E)", fontsize=8)
axs[0].set_title("Entropía Normalizada de VD a LP", fontsize=10)
axs[0].legend(title="Longitud de palabra", fontsize=6)
axs[0].grid()

# Gráfico 2: Entropía Normalizada de LP a VD
for n in range(1, 9):
    axs[1].plot(window_sizes_seconds, entropia_normalizada_lp_vd[f"{n}_bits"], label=f"{n} bits")
axs[1].set_xlabel("Tamaño de ventana (segundos)", fontsize=8)
axs[1].set_ylabel("Entropía Normalizada (E)", fontsize=8)
axs[1].set_title("Entropía Normalizada de LP a VD", fontsize=10)
axs[1].legend(title="Longitud de palabra", fontsize=6)
axs[1].grid()

plt.tight_layout()  # Ajustar espaciado para que los subgráficos no se sobrepongan
plt.show()
```

El análisis de patrones ordinales utiliza la estructura de orden de los valores en series temporales para representar su comportamiento en términos de patrones ordinales. Este enfoque es particularmente útil para el cálculo de la información mutua (IM) en señales EEG y se realiza mediante los siguientes pasos y fórmulas, de acuerdo con el archivo proporcionado:

1. **Definición de Patrones Ordinales**: 
   Dada una serie temporal $ x_t $, los patrones ordinales se definen en función del orden de los valores en segmentos de la serie. Para cada tiempo $ t $, se considera un vector de retardos $ (x_t, x_{t-\tau}, \ldots, x_{t-d\tau}) $, donde $ d $ es el orden del patrón y $ \tau $ es el retardo. El patrón ordinal está determinado por la secuencia de índices que ordena estos valores de mayor a menor, generando una permutación única.

2. **Distribución de Patrones Ordinales**:
   Una vez que se transforman los valores de la serie temporal en patrones ordinales, se calcula la frecuencia relativa de cada patrón en ventanas de tiempo deslizantes de longitud fija $ \delta $. Esto da como resultado una distribución de patrones ordinales $ D_d^\tau(t) $, que se expresa como un vector de frecuencias $ (p_0, p_1, \ldots, p_{(d+1)!-1}) $, donde cada $ p_i $ representa la probabilidad de observar el patrón ordinal $ i $ en la ventana de tiempo.

3. **Entropía de Permutación**:
   La entropía de permutación es una medida de la complejidad de la serie temporal y se calcula a partir de la distribución de patrones ordinales. Para una distribución $ D_d^\tau(t) $, la entropía de permutación $ H_d^\tau(t) $ se define como:
   $ H_d^\tau(t) = -\sum_{i=0}^{(d+1)!-1} p_i \ln p_i $
   Esta medida es robusta frente al ruido y captura la complejidad del sistema que genera la serie temporal.

4. **Longitud de Partes Monótonas**:
   Para analizar la cantidad de comportamiento monótono (creciente o decreciente) en la serie, se definen las métricas $ \text{up}_d^\tau(t) $ y $ \text{down}_d^\tau(t) $, que representan la longitud media de las secuencias monótonamente crecientes y decrecientes, respectivamente. Estas se calculan en función de las frecuencias de los patrones ordinales:
   $ \text{up}_d^\tau(t) = \sum_{l=1}^{d} \sum_{j<\frac{(d+1)!}{l+1}} p_j, \quad \text{down}_d^\tau(t) = \sum_{l=1}^{d} \sum_{j>(d+1)!-1-\frac{(d+1)!}{l+1}} p_j $

La siguiente es una explicación detallada del cálculo de la Información Mutua (IM) ordinal, utilizando las fórmulas involucradas y siguiendo el procedimiento del código proporcionado:

### 1. Cálculo de Patrones Ordinales
El primer paso es transformar cada señal en una secuencia de patrones ordinales de orden $ d $ y retardo $ \tau $. Para una serie temporal $ x_t $, un patrón ordinal de orden $ d $ y retardo $ \tau $ en el tiempo $ t $ se define como una secuencia de índices que representa el orden de los valores $ (x_t, x_{t-\tau}, \ldots, x_{t-d\tau}) $.

Por ejemplo, si tenemos un segmento de datos $ (x_t, x_{t-\tau}, x_{t-2\tau}) $ y su orden es $ x_{t-\tau} < x_{t-2\tau} < x_t $, se asigna la permutación $(1, 2, 0)$, representando el orden relativo de los datos. Este patrón ordinal se repite para toda la serie, generando una lista de patrones ordinales.

### 2. Cálculo de la Distribución de Patrones Ordinales
Una vez obtenida la lista de patrones ordinales, se calcula la frecuencia relativa de cada patrón para obtener su distribución de probabilidades. Si $ N $ es el número total de patrones generados en la señal, y $ n_i $ es la cantidad de veces que aparece el patrón $ i $, la probabilidad de cada patrón $ p_i $ se calcula como:
$$
p_i = \frac{n_i}{N}
$$

### 3. Cálculo de Probabilidad Conjunta de Patrones Ordinales entre dos Señales
Para calcular la información mutua entre dos señales (por ejemplo, LP y VD), primero se obtiene la probabilidad conjunta de los patrones ordinales en ambas señales. Esto se realiza creando pares de patrones simultáneos entre las dos señales y contando su frecuencia.

Si $ N $ es el número total de pares de patrones, y $ n_{ij} $ es la cantidad de veces que el par de patrones $(i, j)$ ocurre, la probabilidad conjunta $ p_{ij} $ se calcula como:
$$
p_{ij} = \frac{n_{ij}}{N}
$$

### 4. Cálculo de la Información Mutua (IM) Ordinal
La información mutua ordinal mide la cantidad de información compartida entre las dos señales, basándose en la probabilidad conjunta y las probabilidades marginales de los patrones ordinales de cada señal. La fórmula para la IM entre las señales $ X $ (LP) y $ Y $ (VD) es:
$$
I(X; Y) = \sum_{i, j} p_{ij} \log \left( \frac{p_{ij}}{p_i p_j} \right)
$$
donde:
- $ p_{ij} $ es la probabilidad conjunta del par de patrones $(i, j)$.
- $ p_i $ y $ p_j $ son las probabilidades marginales de los patrones ordinales en las señales LP y VD, respectivamente.

En el cálculo:
- Para cada par de patrones ordinales $(i, j)$ que tiene una probabilidad conjunta positiva ($ p_{ij} > 0 $), se evalúa el término $ p_{ij} \log \left( \frac{p_{ij}}{p_i p_j} \right) $.
- Este valor se acumula para obtener la información mutumación compartida entre ambas señales.

Este proceso permite medir la dependencia estructural entre las dos señales EEG, utilizando el enfoque de patrones ordinales para una estimación robusta y menos sensible al ruido.

**Análisis Comparativo y Conclusión General:**

El método de **IM usando patrones ordinales** es el más adecuado para los datos EEG debido a su resistencia al ruido y capacidad de capturar dependencias estructurales sin depender de la magnitud. Aunque cada método ofrece ventajas, el análisis de patrones ordinales permite una evaluación precisa de las dependencias en señales EEG.
